```python
import pandas as pd
data=pd.read_csv("students_record.csv")
data["Maths"]=pd.to_numeric(data["Maths"],errors="coerce")
data["Physics"]=pd.to_numeric(data["Physics"],errors="coerce")
data["Chemistry"]=pd.to_numeric(data["Chemistry"],errors="coerce")
data["Attendance"]=pd.to_numeric(data["Attendance"].str.replace("%",""),errors="coerce")
data[["Maths","Physics","Chemistry"]]=data[["Maths","Physics","Chemistry"]].fillna(data[["Maths","Physics","Chemistry"]].mean())
data[["Attendance","Assignments","Quiz"]]=data[["Attendance","Assignments","Quiz"]].fillna(0)
data["Total"]=data["Maths"]+data["Physics"]+data["Chemistry"]
data["Average"]=data["Total"]/3
data["Percentage"]=data["Total"]/300*100
data["PerformanceScore"]=(data["Assignments"]*5)+data["Quiz"]+data["Project"]
data["Efficiency"]=(data["Attendance"]*data["Average"])/100
data.loc[data["Percentage"]>=85,"Grade"]="A"
data.loc[(data["Percentage"]>=70)& (data["Percentage"]<85),"Grade"]="B"
data.loc[(data["Percentage"]>=50)& (data["Percentage"]<70),"Grade"]="C"
data.loc[data["Percentage"]<50,"Grade"]="Fail"
print(data.loc[data["Percentage"].idxmax()])
print(data.loc[data["Percentage"].idxmin()])
print(data.loc[data["Attendance"].idxmax()])
print(data[["Maths","Physics","Chemistry"]].mean())
print(data["Assignments"].mean())
print(data.loc[data["Attendance"]<75])
print(data.loc[data["Attendance"]>70])
print(data.loc[data["Grade"]=="A"])
print(data.loc[(data["Grade"]=="Fail")& (data["Attendance"]>80)])
print(data.nlargest(5,"Percentage"))
print(data.nsmallest(5,"Percentage"))
print(data.nlargest(3,"Efficiency"))
```
